## Who Was Thomas Bayes?

**Thomas Bayes** was an 18th-century English statistician, philosopher, and minister whose posthumously published work laid the foundation for the entire branch of statistics that now bears his name. His revolutionary idea was to formalize a way to incorporate new evidence to update an existing belief.

### The Core of Bayesian Inference: Updating Beliefs

The essence of Bayesian thinking is a natural and intuitive process of learning. Imagine you're deciding whether to take an umbrella when you leave the house.

1.  **Initial Belief**: You hear a weather forecast predicting a 50% chance of rain. You have a moderate belief that it might rain.
2.  **New Evidence**: You look out the window and see dark clouds and rain beginning to fall.
3.  **Updated Belief**: You update your belief based on this new, direct evidence. You are now almost certain it will rain and decide to take the umbrella.

You have just performed an intuitive Bayesian analysis. **Bayesian inference** is the formal process of updating one's belief about an unknown quantity (like a model parameter) as new information (data) becomes available.

### Bayesian vs. Frequentist (Classical) Statistics

The Bayesian approach differs from the more traditional frequentist approach in two profound ways, which are summarized in the table below.

| Concept | Frequentist (Classical) Approach | Bayesian Approach |
| :--- | :--- | :--- |
| **Probability** | The long-run proportion of outcomes. | A degree of belief or confidence in a statement. |
| **Parameters** | Fixed, unknown constants. | Random variables described by distributions. |

### The Meaning of Probability

  * **Frequentist View**: Probability is defined as the long-run frequency of an event. To say the probability of a die landing on six is $1/6$ means that if you were to roll the die thousands of times, you would expect approximately one-sixth of the outcomes to be six. 
  * **Bayesian View**: Probability represents a **degree of belief** about a proposition. This allows for probabilistic statements about things that are not long-run events, such as a model parameter. A Bayesian might say, "I am 90% certain that the average effect of this treatment is positive." 

### The Nature of Model Parameters

  * **Frequentist View**: A population parameter, such as the true mean ($\\mu$), is considered a **fixed, unknown constant**. Our goal is to use sample data to calculate a point estimate (like the sample mean, $\\bar{x}$) and a confidence interval that hopefully captures this single true value.
  * **Bayesian View**: A parameter is considered a **random variable**. Because it is a random variable, we can describe our uncertainty about it using a **probability distribution**. Instead of a single point estimate, the result of a Bayesian analysis is a full probability distribution for the parameter, which tells us the entire range of plausible values and their associated probabilities. 

### The Advantages of the Bayesian Approach

These philosophical differences lead to several practical advantages:

  * **Natural Handling of Uncertainty**: Since parameters are represented by distributions, our uncertainty is captured directly. A wide distribution for a parameter indicates high uncertainty, while a narrow distribution indicates high certainty.
  * **Incorporation of Prior Knowledge**: The "degree of belief" interpretation of probability allows for the formal inclusion of expert opinion or prior knowledge into a model in a statistically sound way.
  * **Intuitive Results**: The results (like a 95% credible interval) have a more direct interpretation than their frequentist counterparts.
  * **Statistical Coherence**: The methods are statistically valid even with small sample sizes.


### The Role of Probability Distributions

Probability distributions are central to the Bayesian paradigm. A distribution specifies the possible values a random variable can take and their corresponding probabilities. In Python, a common way to work with a distribution is to generate a large number of **draws** (or samples) from it. The frequency of values in these draws reflects the underlying probability distribution.

We can visualize these draws using a Kernel Density Estimate (KDE) plot, which creates a smooth curve representing the distribution's shape.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate 10,000 "draws" from a normal distribution
# This represents a probability distribution for a parameter.
np.random.seed(42)
draws = np.random.normal(loc=170, scale=10, size=10000)

# Visualize the distribution using the OOP approach
fig, ax = plt.subplots(figsize=(10, 6))

sns.kdeplot(x=draws, fill=True, ax=ax)

ax.set_title("Density Plot of a Probability Distribution")
ax.set_xlabel("Parameter Value")
ax.set_ylabel("Density")
ax.grid(True, linestyle='--')
plt.show()
```

## The Foundations of Probability

Probability theory quantifies uncertainty. A probability is a number in $[0,1]$, where 0 means impossible and 1 means certain.

### Basic Probability Rules

Let $A$ and $B$ be events.  
- **Mutually exclusive** means $A \cap B = \emptyset$.  
- **Independent** means $\Pr(A \cap B) = \Pr(A)\Pr(B)$.

#### Sum Rule (OR / Union)
- If $A$ and $B$ are mutually exclusive:
  $$
  \Pr(A \cup B) = \Pr(A) + \Pr(B)
  $$
- General form (always true):
  $$
  \Pr(A \cup B) = \Pr(A) + \Pr(B) - \Pr(A \cap B)
  $$
- **Example**: Rolling a 2 or a 4 on a fair die:
  $$
  \Pr(\{2\} \cup \{4\}) = \Pr(2) + \Pr(4) = \frac{1}{6} + \frac{1}{6} = \frac{1}{3} \approx 0.\overline{3}
  $$

#### Product Rule (AND / Intersection, for Independent Events)
- If $A$ and $B$ are independent:
  $$
  \Pr(A \cap B) = \Pr(A)\Pr(B)
  $$
- **Example**: Rolling a 2 *and then* a 4 on two independent fair rolls:
  $$
  \Pr(2 \text{ then } 4) = \Pr(2)\Pr(4) = \frac{1}{6} \cdot \frac{1}{6} = \frac{1}{36} \approx 0.0278
  $$

### Conditional Probability

The conditional probability of $A$ given $B$ (with $\Pr(B) > 0$) is:
$$
\Pr(A \mid B) = \frac{\Pr(A \cap B)}{\Pr(B)}
$$

**Example**: Urn with two orange and one blue ball (no replacement):
- First draw (unconditional): $\Pr(\text{orange}) = \frac{2}{3}$
- Second draw after removing an orange: $\Pr(\text{blue} \mid \text{orange first}) = \frac{1}{2}$

### Bayes' Theorem

Bayes' Theorem inverts conditionals:
$$
\Pr(A \mid B) = \frac{\Pr(B \mid A)\Pr(A)}{\Pr(B)}
$$
Paired with the Law of Total Probability:
$$
\Pr(B) = \Pr(B \mid A)\Pr(A) + \Pr(B \mid \neg A)\Pr(\neg A)
$$

### Worked Example: Bayes' Theorem in Practice

Probability of an accident given a slippery road:
$$
\Pr(\text{accident} \mid \text{slippery}) = \frac{\Pr(\text{slippery} \mid \text{accident}) \cdot \Pr(\text{accident})}{\Pr(\text{slippery})}
$$

```python
import numpy as np
import pandas as pd

def bayes_demo(n_trips=10000,
               seed=42,
               p_accident=0.06,
               p_slippery_given_accident=0.70,
               p_slippery_given_no_accident=0.03):
    """
    Simulate road-trip data to demonstrate Bayes' Theorem and verify it empirically.

    The simulation generates a Boolean series for accidents and a dependent Boolean series
    for slippery conditions. The dependence is controlled by the conditional probabilities
    p_slippery_given_accident and p_slippery_given_no_accident. Using the simulated data,
    the function computes:
        - Prior:            P(accident)
        - Evidence:         P(slippery)
        - Likelihood:       P(slippery | accident)
        - Posterior (via Bayes):      P(accident | slippery) computed by the formula
        - Posterior (empirical):      P(accident | slippery) computed by direct conditioning
        - Analytical baseline:        Posterior computed from the model parameters via
                                      the law of total probability

    Parameters
    ----------
    n_trips : int
        Number of simulated trips.
    seed : int
        Seed for the pseudo-random number generator for reproducibility.
    p_accident : float
        Prior probability of an accident on any given trip.
    p_slippery_given_accident : float
        Probability the road is slippery given an accident occurred.
    p_slippery_given_no_accident : float
        Probability the road is slippery given no accident occurred.

    Returns
    -------
    pd.Series
        A labelled collection of probabilities for convenient display and downstream use.
    """
    rng = np.random.default_rng(seed)

    # Simulate accidents as Bernoulli(p_accident).
    accidents = rng.random(n_trips) < p_accident

    # Simulate slippery conditions with probabilities conditional on 'accidents'.
    # Use a single uniform array for efficiency and deterministic alignment.
    u = rng.random(n_trips)
    slippery = np.empty(n_trips, dtype=bool)
    slippery[accidents] = u[accidents] < p_slippery_given_accident
    slippery[~accidents] = u[~accidents] < p_slippery_given_no_accident

    # Wrap into a DataFrame for clarity and potential exploration.
    df = pd.DataFrame({"accident": accidents, "slippery": slippery})

    # Components of Bayes theorem estimated from the simulation.
    p_A = df["accident"].mean()                                   # P(A)
    p_B = df["slippery"].mean()                                   # P(B)
    p_B_given_A = df.loc[df["accident"], "slippery"].mean()       # P(B|A)

    # Posterior computed via Bayes' formula using simulated components.
    p_A_given_B_formula = (p_B_given_A * p_A) / p_B

    # Posterior computed directly from the simulated joint by conditioning.
    p_A_given_B_empirical = df.loc[df["slippery"], "accident"].mean()

    # Analytical posterior from the model parameters (sanity check).
    p_B_analytical = (p_slippery_given_accident * p_accident
                      + p_slippery_given_no_accident * (1.0 - p_accident))
    p_A_given_B_analytical = (p_slippery_given_accident * p_accident) / p_B_analytical

    # Assemble results.
    out = pd.Series({
        "P(accident) - Prior": p_A,
        "P(slippery) - Evidence": p_B,
        "P(slippery | accident) - Likelihood": p_B_given_A,
        "P(accident | slippery) - Posterior (Bayes formula)": p_A_given_B_formula,
        "P(accident | slippery) - Posterior (empirical)": p_A_given_B_empirical,
        "P(accident | slippery) - Posterior (analytical)": p_A_given_B_analytical
    })

    return out

# Run the demonstration and print results with consistent formatting.
results = bayes_demo()
for k, v in results.items():
    print(f"{k:55s} {v:.4f}")
